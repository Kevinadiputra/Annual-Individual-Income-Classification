#Prediksi pendapatan Individu
### Dataset yang Digunakan: Adult Dataset  
Dataset ini berasal dari UCI Machine Learning Repository dan dapat diakses melalui tautan berikut:  
[Adult Dataset](https://archive.ics.uci.edu/dataset/2/adult)  

---

### **Deskripsi Dataset**  
Dataset ini terdiri dari **14 atribut** dan **1 kolom label**, di mana label memiliki **2 kelas**. Atribut-atribut tersebut mencakup berbagai informasi demografis dan ekonomi yang digunakan untuk memprediksi apakah pendapatan tahunan seseorang melebihi **$50,000 USD** berdasarkan data sensus.

#3. Preprocessing

```python
# @title mengubah isi dari kolom income menjadi binary
df_adult['income'] = df_adult['income'].replace({'<=50K': 0, '>50K': 1})
df_adult['income'] = df_adult['income'].replace({'<=50K.': 0, '>50K.': 1})
```

```python
print(df_adult['income'].unique())
```

```python
# @title one hot encoding
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
df = df_adult.copy()

# Ordinal Encoding
ordinal_features = ['education', 'marital-status']
education_order = ['Preschool', '1st-4th', '5th-6th', '7th-8th', '9th', '10th', '11th', '12th',
                   'HS-grad', 'Some-college', 'Assoc-voc', 'Assoc-acdm', 'Bachelors',
                   'Masters', 'Prof-school', 'Doctorate']
marital_status_order = ['Never-married', 'Separated', 'Divorced', 'Widowed',
                        'Married-spouse-absent', 'Married-AF-spouse', 'Married-civ-spouse']

ordinal_encoder = OrdinalEncoder(categories=[education_order, marital_status_order])
df[ordinal_features] = ordinal_encoder.fit_transform(df[ordinal_features])

# One-Hot Encoding for Nominal Attributes
nominal_features = ['workclass', 'occupation', 'relationship', 'race', 'sex', 'native-country']
df_nominal = pd.get_dummies(df[nominal_features], drop_first=True)

# Gabungkan hasil encoding
df_encoded = pd.concat([df.drop(columns=nominal_features), df_nominal], axis=1)

# mengubah tipe datannya menjadi float
df_encoded = df_encoded.astype(float)
```

```python
# @title menghapus outlier
def remove_outliers(df, columns):
    # Menyimpan dataframe hasil tanpa outlier
    df_resampled = df.copy()

    for col in columns:
        # Menghitung IQR (Interquartile Range)
        Q1 = df_resampled[col].quantile(0.25)
        Q3 = df_resampled[col].quantile(0.75)
        IQR = Q3 - Q1

        # Menentukan batas bawah dan atas untuk outlier
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Menghapus baris yang memiliki outlier
        df_resampled = df_resampled[(df_resampled[col] >= lower_bound) & (df_resampled[col] <= upper_bound)]

    return df_resampled

# Menghapus outlier pada dataframe df_adult
df_resampled_cleaned = remove_outliers(df_resampled, numeric_columns)

# Menampilkan hasil
print(f"Dataframe setelah menghapus outlier: {df_resampled_cleaned.shape}")
```

#5. Model

```python
# @title memisahkan atribut independen dan dependen data yang tidak dihapus outliernnya
x = df_resampled.drop(['income'], axis=1)
y = df_resampled['income']
```

```python
# @title memisahkan atribut independen dan dependen data yang dihapus outliernnya
x_noout = df_resampled_cleaned.drop(['income'], axis=1)
y_noout = df_resampled_cleaned['income']
```

```python
# @title lakukan normalisasi pada data independen yang tidak dihapus outliernnya
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
```

```python
# @title lakukan normalisasi pada data independen yang dihapus outliernnya
x_scaled_noout = scaler.fit_transform(x_noout)
```

```python
# @title train test split data yang tidak dihapus outliernnya
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

```python
# @title train test split data yang dihapus outliernnya
x_train_noout, x_test_noout, y_train_noout, y_test_noout = train_test_split(x_scaled_noout, y_noout, test_size=0.2, random_state=42)
```

```python
# melihat bentuk data
print(f'x_train: {x_train.shape}')
print(f'x_test: {x_test.shape}')
print(f'y_train: {y_train.shape}')
print(f'y_test: {y_test.shape}')
```

```python
# @title latih model menggunakan data yang outliernnya tidak dihapus
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import cross_val_score

# Inisialisasi model
models = {
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Support Vector Machine (SVM)": SVC(),
    "Logistic Regression": LogisticRegression(),
    "Naive Bayes": GaussianNB(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "XGBoost": XGBClassifier(),
    "AdaBoost": AdaBoostClassifier()
}

# Menyimpan hasil evaluasi
results = []
predictions = {}

for name, model in models.items():
    print(f"\n=== {name} ===")

    # Melatih model
    model.fit(x_train, y_train)

    # Prediksi pada data uji
    y_pred = model.predict(x_test)

    # Menyimpan prediksi dan aktual
    predictions[name] = pd.DataFrame({
        "Actual": y_test,
        "Predicted": y_pred
    })

    # Menghitung akurasi
    acc = accuracy_score(y_test, y_pred)
    print(f"Akurasi: {acc:.2f}")

    # Menampilkan classification report
    print("Classification Report:")
    print(classification_report(y_test, y_pred))

    # Menyimpan hasil untuk analisis lebih lanjut
    results.append({"Model": name, "Accuracy": acc})

    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)

    # Visualisasi confusion matrix
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=model.classes_, yticklabels=model.classes_)
    plt.title(f"Confusion Matrix: {name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

# Menyusun hasil ke dalam DataFrame
results_df = pd.DataFrame(results)
print("\nRingkasan Hasil:")
print(results_df)

# Menampilkan prediksi dan aktual untuk tiap model
for model_name, df in predictions.items():
    print(f"\n=== Hasil Prediksi dan Aktual: {model_name} ===")
    print(df.head())
```

Sehingga disini didapatkan bahwa data yang outliernnya tidak dihapus memiliki akurasi lebih baik, terutama pada model XGBoost, sehingga pada kasus kali ini saya akan menggunakan XGBoost dengan data yang tidak di hapus outliernnya.

```python
# @title mencoba hyperparameter tunning XGBoost
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier

# Inisialisasi model XGBoost
xgb_model = XGBClassifier(eval_metric='mlogloss')

# Tentukan grid hyperparameter yang ingin dicoba
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.8, 0.9, 1.0],
    'gamma': [0, 0.1, 0.2],
}

# Inisialisasi GridSearchCV
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid,
                           scoring='accuracy', cv=5, verbose=1, n_jobs=-1)

# Melakukan pencarian grid pada data training
grid_search.fit(x_train, y_train)

# Menampilkan hasil terbaik dari pencarian hyperparameter
print("Best Hyperparameters: ", grid_search.best_params_)
print("Best Accuracy: ", grid_search.best_score_)
best_model = grid_search.best_estimator_
y_pred_best = best_model.predict(x_test)
from sklearn.metrics import accuracy_score, classification_report
acc_best = accuracy_score(y_test, y_pred_best)
print(f"Akurasi Model Terbaik: {acc_best:.2f}")
print("Classification Report:")
print(classification_report(y_test, y_pred_best))
```

Ternyata setelah dilakukan hyperparameter tunning, model hanya mengalami sedikit peningkatan akurasi

```python
# @title visualisasi  confusion matrix
cm_best = confusion_matrix(y_test, y_pred_best)

# Visualisasi confusion matrix
plt.figure(figsize=(6, 4))
sns.heatmap(cm_best, annot=True, fmt='d', cmap='Blues', xticklabels=best_model.classes_, yticklabels=best_model.classes_)
plt.title("Confusion Matrix: Best Model")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
